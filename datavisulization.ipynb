```python
import matplotlib.pyplot as plt
import numpy as np
##Draw a line in a diagram from position (0,10) to position (5,150)
xpoints=np.array([1,2])
ypoints=np.array([6,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([6,10])
ypoints=np.array([6,10])
plt.plot(xpoints,ypoints,'*')##put x on the points we can change a symbol
plt.show()
```

```python
xpoints=np.array([6,10,20,45,78])
ypoints=np.array([6,15,19,23,25])
plt.plot(xpoints,ypoints,marker="*",ms=20,mfc="y",mec="r")##put x on the points we can change a symbol
plt.show()
```

```python
ypoints=np.array([6,10,12,15,45])
plt.plot(ypoints,'*'"r")
plt.show()
```

```python
ypoints=np.array([3,8,1,10,12,16])
plt.plot(ypoints,linestyle='dashed',marker='o',ms=20,mec="black",color="yellow",linewidth=5)
plt.show()
```

```python
y1=np.array([2,24,16,18,10])
y2=np.array([3,6,10,22,15])
plt.plot(y1,color="red")
plt.plot(y2,'g')
plt.show()
```

```python
y1=np.array([2,24,16,18,10])
y2=np.array([3,6,10,22,15])
plt.plot(y1,y2,marker='*',ms=15,mfc='red',linewidth=5)
plt.title("Sports watch data",fontsize=20,color='black')
plt.xlabel("Average Pulse",fontsize=20,color="red",loc='center')
plt.ylabel("Normal Pulse",fontsize=20,color="yellow")
plt.grid(color="green", linestyle="dashed")
plt.show()
```

```python
y1=np.array([2,24,16,18,10])
y2=np.array([3,6,10,22,15])
plt.scatter(y1,y2)#find out the relation between two numarical attribute
```

```python
y1=np.array(['A','B','C','D','E'])
y2=np.array([3,6,10,22,15])
plt.bar(y1,y2,color="orange",width=0.2)
plt.show()
```

```python
y1=np.array(['A','B','C','D','E'])
y2=np.array([3,6,10,22,15])
plt.barh(y1,y2,color="orange")
plt.show()
```

```python
y=np.array([15,150,20,14])
plt.pie(y)
plt.show()
```

```python
y=np.array([15,150,20,14])
mylabel=["apple","mango","banna","Orange"]
plt.pie(y,labels=mylabel)
plt.show()
```

```python
y=np.array([15,150,20,14])
mylabel=["apple","mango","banna","Orange"]
plt.pie(y,labels=mylabel,startangle=180)
plt.legend(labels=mylabel)
plt.show()
```

```python
y=np.array([15,150,20,14])
mylabel=["apple","mango","banna","Orange"]
myexplods=[0,0.95,0.8,0.2]
mycolor=["red","orange","yellow","hotpink"]
plt.pie(y,labels=mylabel,explode=myexplods,colors=mycolor)
plt.show()
```
